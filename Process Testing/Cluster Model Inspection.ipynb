# Cluster Model Inspection
This notebook contains some exploratory analysis of the results from the various clustering methods we tried in the patient_clusters.py script.

We're using a Jupyter Notebook here rather than just a script because the code itself will be simple, and the process is more narrative. Later in model development we will aim to formalise some of the inspection process to permit the choice of clustering method through user-defined parameters.

```python
import pandas as pd
import matplotlib.pyplot as plt
robust = pd.read_csv("robust_scaler_clustering.csv")
standard = pd.read_csv("standard_scaler_clustering.csv")
```

Let's compare the results from each algorithm in turn.

```python
display(robust.loc[robust["algorithm"] == "gmm"])
display(standard.loc[standard["algorithm"] == "gmm"])
```

```python
display(robust.loc[robust["algorithm"] == "kmeans"])
display(standard.loc[standard["algorithm"] == "kmeans"])
```

```python
display(robust.loc[robust["algorithm"] == "agglomerative"])
display(standard.loc[standard["algorithm"] == "agglomerative"])
```

```python
display(robust.loc[robust["algorithm"] == "DBSCAN"])
display(standard.loc[standard["algorithm"] == "DBSCAN"])
```

We can see the greatest difference between the two scaling options is that the RobustScaler() givese much higher silhouette scores, suggesting better-defined clusters where points are similar to others in their cluster. Although the StandardScaler() does typically give slightly lower SSE, combined with the low silhouette scores this suggests that although the clusters formed are tighter, they are somewhat more arbitrary. This may be due to the fact that the StandardScaler() method fails to preserve the shape of the underlying distributions, in contrast to RobustScaler().

Two algorithms stand out as the best performing when considering both metrics: k-Means and Agglomerative. Let's inspect them closer.

```python
fig, ax = plt.subplots(1, 2, figsize=(20,8))

# Plot k-means SSE and Silhouette score
ax[0].plot(robust.loc[robust["algorithm"] == "kmeans","num_clusters"], robust.loc[robust["algorithm"] == "kmeans","sse_scores"], label='SSE', color='blue')
ax0_twin = ax[0].twinx()
ax0_twin.plot(robust.loc[robust["algorithm"] == "kmeans","num_clusters"], robust.loc[robust["algorithm"] == "kmeans","silhouette_scores"], label='Silhouette Score', color="red")
ax[0].set_title("k-Means")
ax[0].set_xlabel("Num. Clusters")
ax[0].set_ylabel("SSE")
ax0_twin.set_ylabel('Silhouette Score')

# Plot agglomerative SSE and Silhouette score
ax[1].plot(robust.loc[robust["algorithm"] == "agglomerative","num_clusters"], robust.loc[robust["algorithm"] == "agglomerative","sse_scores"], label='SSE', color='blue')
ax1_twin = ax[1].twinx()
ax1_twin.plot(robust.loc[robust["algorithm"] == "agglomerative","num_clusters"], robust.loc[robust["algorithm"] == "agglomerative","silhouette_scores"], label='Silhouette Score', color="red")
ax[1].set_title("Agglomerative")
ax[1].set_xlabel("Num. Clusters")
ax[1].set_ylabel("SSE")
ax1_twin.set_ylabel('Silhouette Score')

# Set y-axis limits
ymin = [0, 0]
ymax = [1.1*robust.loc[robust["algorithm"].isin(["kmeans", "agglomerative"]), "sse_scores"].max(), 1]
ax[0].set_ylim([ymin[0], ymax[0]])
ax0_twin.set_ylim([ymin[1], ymax[1]])
ax[1].set_ylim([ymin[0], ymax[0]])
ax1_twin.set_ylim([ymin[1], ymax[1]])

# Add legend using lines from axis and twin axis.
handles1, labels1 = ax[0].get_legend_handles_labels()
handles2, labels2 = ax0_twin.get_legend_handles_labels()
handles = handles1 + handles2
labels = labels1 + labels2
fig.legend(handles, labels, loc='upper center', ncol=2)
```

Both algorithms show an "elbow" in the SSE at 4 clusters, after which there are diminishing returns on the reduction in error. In the k-means case there is a slight drop in silhouette score compared to 3 clusters, whereas the agglomerative algorithm has a local peak. This suggests that in both cases, 4 clusters is likely the optimal number.

The overall results are quite similar, so k-Means may be the best "default" choice for the model just based on its simplicity of use and quick runtime even for very large datasets. However, it is worth considering other options and potentially including them as model options later in development.
