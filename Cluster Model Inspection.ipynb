# Cluster Model Inspection
This notebook contains some exploratory analysis of the results from the various clustering methods we tried in the patient_clusters.py script.

We're using a Jupyter Notebook here rather than just a script because the code itself will be simple, and the process is more narrative. Later in model development we will aim to formalise some of the inspection process to permit the choice of clustering method through user-defined parameters.

```python
import pandas as pd
import matplotlib.pyplot as plt
robust = pd.read_csv("robust_scaler_clustering.csv")
standard = pd.read_csv("standard_scaler_clustering.csv")
```

Let's compare the results from each algorithm in turn.

```python
display(robust.loc[robust["algorithm"] == "gmm"])
display(standard.loc[standard["algorithm"] == "gmm"])
```

```python
display(robust.loc[robust["algorithm"] == "kmeans"])
display(standard.loc[standard["algorithm"] == "kmeans"])
```

```python
display(robust.loc[robust["algorithm"] == "agglomerative"])
display(standard.loc[standard["algorithm"] == "agglomerative"])
```

```python
display(robust.loc[robust["algorithm"] == "DBSCAN"])
display(standard.loc[standard["algorithm"] == "DBSCAN"])
```

So we can see in all cases that the sse_scores are broadly similar, both in values and trends. This tells us that the optimal number of clusters suggested by each algorithm is the same regardless of the scaling. However, the RobustScaler() data tends to have a higher silhouette score, suggesting better-defined clusters where points are similar to others in their cluster.

Hence, it looks like the RobustScaler() generally performs better. This is likely due to StandardScaler() failing to preserve
the shape of the underlying distributions, and its sensitivity to outliers (particularly in high-variance fields like
LoS), while RobustScaler() avoids both these problems.

Two algorithms stand out as the best performing in terms of low SSE and high silhouette score: k-Means and Agglomerative. Let's inspect them closer.

```python
fig, ax = plt.subplots(1, 2, figsize=(10,8))
ax[0].plot(robust.loc[robust["algorithm"] == "kmeans","num_clusters"], robust.loc[robust["algorithm"] == "kmeans","sse_scores"])
ax[0].set_title("k-Means")
ax[0].set_xlabel("Num. Clusters")
ax[0].set_ylabel("SSE")
ax[1].plot(robust.loc[robust["algorithm"] == "agglomerative","num_clusters"], robust.loc[robust["algorithm"] == "agglomerative","sse_scores"])
ax[1].set_title("Agglomerative")
ax[1].set_xlabel("Num. Clusters")
ax[1].set_ylabel("SSE")
```

The Agglomerative algorithm shows a relatively clear "elbow" at 3 clusters, after which there are diminishing returns on the reduction in error. The k-Means algorthm is less clear, and arguments could be made for between 3-5 clusters being optimal for this particular dataset. However, looking at the steep drop-off in silhouette score at 5 clusters, 3 or 4 is likely the best choice for k-Means here.

The overall results are broadly similar, so k-Means may be the best "default" choice for the model just based on its simplicity of use and quick runtime even for very large datasets.
